```python
print(df["mainroad"].value_counts(""))
print(df["guestroom"].value_counts(""))
print(df["basement"].value_counts(""))
print(df["hotwaterheating"].value_counts(""))
print(df["airconditioning"].value_counts(""))
print(df["prefarea"].value_counts(""))
print(df["furnishingstatus"].value_counts(""))
```

```python
fig,ax = plt.subplots(2,3)
ax[0,0].bar(df["mainroad"].value_counts("").sort_index().index,df["mainroad"].value_counts("").sort_index().values,color = "green")
ax[0,0].set_xticks(df["mainroad"].value_counts("").sort_index().index)
ax[0,0].set_title("mainroad")
ax[0,0].set_ylabel("Count")
ax[0,0].set_xlabel("Values")

ax[0,1].bar(df["guestroom"].value_counts("").sort_index().index,df["guestroom"].value_counts("").sort_index().values,color = "blue")
ax[0,1].set_xticks(df["guestroom"].value_counts("").sort_index().index)
ax[0,1].set_title("guestroom")
ax[0,1].set_ylabel("Count")
ax[0,1].set_xlabel("Values")

ax[0,2].bar(df["basement"].value_counts("").sort_index().index,df["basement"].value_counts("").sort_index().values,color = "orange")
ax[0,2].set_xticks(df["basement"].value_counts("").sort_index().index)
ax[0,2].set_title("basement")
ax[0,2].set_ylabel("Count")
ax[0,2].set_xlabel("Values")

ax[1,0].bar(df["hotwaterheating"].value_counts("").sort_index().index,df["hotwaterheating"].value_counts("").sort_index().values)
ax[1,0].set_xticks(df["hotwaterheating"].value_counts("").sort_index().index)
ax[1,0].set_title("hotwaterheating")
ax[1,0].set_ylabel("Count")
ax[1,0].set_xlabel("Values")

ax[1,1].bar(df["airconditioning"].value_counts("").sort_index().index,df["airconditioning"].value_counts("").sort_index().values)
ax[1,1].set_xticks(df["airconditioning"].value_counts("").sort_index().index)
ax[1,1].set_title("airconditioning")
ax[1,1].set_ylabel("Count")
ax[1,1].set_xlabel("Values")

ax[1,2].bar(df["prefarea"].value_counts("").sort_index().index,df["prefarea"].value_counts("").sort_index().values)
ax[1,2].set_xticks(df["prefarea"].value_counts("").sort_index().index)
ax[1,2].set_title("prefarea")
ax[1,2].set_ylabel("Count")
ax[1,2].set_xlabel("Values")
plt.tight_layout()
plt.show()

plt.bar(df["furnishingstatus"].value_counts("").sort_index().index,df["furnishingstatus"].value_counts("").sort_index().values)
plt.xticks(df["furnishingstatus"].value_counts("").sort_index().index)
plt.title("furnishingstatus")
plt.ylabel("Count")
plt.xlabel("Values")
plt.show()


```

This analysis shows how the categorical features are distributed across the housing dataset.
1) Main road: About 86% of the houses are located on a main road, so this feature is very common and not a strong differentiator.
2) Guest room: Only 18% of houses have a guest room, making it a relatively rare and premium feature.
3) Basement: Around 35% of houses include a basement; most houses do not have one.
4) Hot water heating: Present in only 5% of houses, indicating an extremely rare facility.
5) Air conditioning: Available in approximately 32% of houses, showing moderate adoption.
6) Preferred area: Only 23% of houses are in preferred areas, so such locations are uncommon.
7) Furnishing status:
     42% are semi-furnished
     32% are unfurnished
     26% are fully furnished

From the above analysis, we can conclude that the dataset mainly represents average or middle-range houses rather than luxury houses. Most houses have common features like main road access and are semi-furnished, while facilities such as guest rooms, basements, hot water heating, air conditioning, and preferred area locations are available only in a smaller number of houses.

The categorical data is not evenly distributed, as some categories dominate over others. This shows that certain features are common across most houses, whereas some features are rare and may indicate higher value. Overall, the dataset seems realistic and suitable for further analysis, but care should be taken while modeling because imbalanced categories can affect the results.

1) Bedrooms:

2) athrooms:

4) Parking:

From the above analysis, we can conclude that the dataset mainly represents standard residential housing, suitable for small to medium-sized families. Most houses have moderate numbers of bedrooms, bathrooms, floors, and limited parking facilities, while larger houses with more rooms, bathrooms, or parking space are uncommon.

The numerical features are also unevenly distributed, with most data concentrated around lower to mid-range values. This suggests that the dataset reflects realistic housing conditions, but special care should be taken during modeling since extreme values are rare and may disproportionately influence predictions.
